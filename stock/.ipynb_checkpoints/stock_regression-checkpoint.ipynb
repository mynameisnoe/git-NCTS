```python
"""
Created on Tue Apr  9 22:03:50 2019

@author: noe
"""

import pandas as pd
import matplotlib.pyplot as plt

# 預測台積電隔天收盤價格
data = pd.read_csv("2330.TW.csv")
data = data.dropna()
data = data.iloc[:,4].values

# 用來留30天當test data
periods = 30

x_train = data[0:(len(data) - (len(data) % periods))]
x_train = x_train.reshape(-1, 1)  


# y為x要regression的下一天
y_train = data[1:(len(data) - (len(data) % periods)) + 1] 
y_train = y_train.reshape(-1, 1)


x_test = data[-(periods + 1):]
# number to 1220列
x_test = x_test[:periods]
x_test = x_test.reshape(-1, 1)

# number to 1221列
y_test = data[-(periods):]
y_test = y_test.reshape(-1, 1)

from sklearn.ensemble import RandomForestRegressor
Regressor = RandomForestRegressor()
Regressor.fit(x_train, y_train)

y_pred = Regressor.predict(x_test)

plt.plot(y_test, '*', markersize=5, label = 'Real_Value')
plt.plot(y_pred, 'o', markersize=5, label = 'Predict_Value')
plt.legend()

plt.plot(y_test, label = 'Real_value')
plt.plot(y_pred, label = 'Predict_Value')
plt.legend()
```
