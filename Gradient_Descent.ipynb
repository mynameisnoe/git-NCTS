```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

%matplotlib inline
```

```python

def show_graph(lr, epochs, x_start):
    
    def function(x):
        return np.cos(x)
    def dfunction(x):
        return -np.sin(x)
    
    def Gradient(lr, epochs, x_start, df=dfunction):
    
        x_store = np.zeros(epochs+1)    
        x = x_start    
        x_store[0] = x
        
        for i in range(epochs):         
            
            x_differentiate = df(x)        
            
            gradient = - x_differentiate * lr        
            x += gradient        
            x_store[i+1] = x   
            
        return x_store
    
    x = Gradient(lr, epochs, x_start) 
    
    from numpy import arange
    time = arange(-10.0, 10.0, 0.01)
    plt.plot(time, function(time), c='k')
    
    plt.plot(x, function(x), c='r', label='learning_rate={}'.format(lr))    
    plt.scatter(x, function(x), c='m', s=20.0)   
    
    plt.legend()
    plt.show()
    
```

```python
show_graph(lr=0.3, epochs=10, x_start=1)
```

```python
interact(show_graph, lr=(0, 1, 0.1), epochs=(10, 50, 10), x_start=(-10,10, 0.5))
```
